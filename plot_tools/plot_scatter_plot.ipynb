```python
data_std = []
for model_data in mv_data:
    data_std.append({"model": model_data["model"],
                     "image_tail": np.std(model_data["image_tail"]), "image_noise": np.std(model_data["image_noise"]),
                     "pixel_tail": np.std(model_data["pixel_tail"]), "pixel_noise": np.std(model_data["pixel_noise"])})
```
